# Applying the Association Rules


```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

Creation of Frequent Itemssets

```python
freq_items= apriori(df,min_support =0.1,use_colnames=True) # selecting the thershold value as 0.1  to create a frequent item sets
freq_items
```

Creation of Rules using Frequent itme sets
selecting the diffrent diffrent metric ==> lift and confidence

```python
rules =association_rules(freq_items, metric ="confidence",min_threshold=0.7)# we selected the metric is confidence above .07
rules
```

if the metric is confidence" and the threshold values is "0.7"
the antecedent with (RefBks or artbooks with remaining books) are more selling
And with consequent as cookbooks and child books as consequent

```python
rules1 = association_rules(freq_items ,metric="lift", min_threshold =2.2)
rules1
```

[The metric as "lift" with the minimum threshold value is 2.2 the most selling books are====
Cookbooks, ArtBooks  GeographyBooks, And Child Books with lift metric is greater than "2.2]"

Customising the Association Rules  
1. atleast 2 anticedent Rules
2. whose confidence value is greater than > 0.75 or 0.80
3. with lift ratio above 2 or 2.1

```python
# computing the length of the antcedents
rules['antecedents_len'] = rules['antecedents'].apply(lambda x: len(x))
rules
```

```python
customised_rules =rules[ (rules['antecedents_len'] >= 2) &
(rules['confidence'] > 0.8)&
(rules['lift'] > 1.5)]
```

```python
# if lift = 1 the probablity of occurence of antecdent and consequent is independent to each other
# if lift > 1 the proobablity of coccurence  degree of two items are dependent to each other
# if lift < 1 it tells that one item is subsute of another item, whcih means item has a negitive effect on each other
```
