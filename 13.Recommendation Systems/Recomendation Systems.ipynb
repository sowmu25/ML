# Calculation of Cosine  and Correlation Similarity Beteween Users

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
similarity_books =1 - pairwise_distances(book_df,metric ="cosine")
```
